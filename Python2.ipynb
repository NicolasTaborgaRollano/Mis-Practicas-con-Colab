```python
import pretty_midi
from IPython.display import Audio
midi = pretty_midi.PrettyMIDI()
instrument = pretty_midi.Instrument(program=0)

for i in range(60,72):
    note = pretty_midi.Note(
        velocity=100,
        pitch=i,
        start=(i-60)*0.5,
        end=(i-60)*0.5 + 0.4

    )
instrument.notes.append(note)
midi.instruments.append(instrument)
audio = midi.fluidsynth()
Audio(audio, rate=44100)
```

```python
import tarfile

with open ("archivo1.txt","w") as f:
  f.write("Hola mundo")

with open ("archivo2.txt", "w") as f:
  f.write("Archivo de prueba")

with tarfile.open("datos.tar.gz","w:gz") as tar:
  tar.add("archivo1.txt")
  tar.add("archivo2.txt")

print("Archivo comprimido creado")
```
